# PCA

用鳶尾花練習

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
from sklearn.decomposition import PCA
```

## step 1. 開 PCA 機

```python
pca = PCA(n_components=2)
```

## step 2. 訓練

```python
pca.fit(x)
```

## step 3. 預測

```python
x[87]
```

```python
X = pca.transform(x)
```

```python
# 真的變二維了！
X[87]
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

### 準備訓練/測試資料

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=0)
```

### 1. 開 SVC 函數學習機

```python
from sklearn.svm import SVC
```

```python
clf = SVC(gamma='auto')
```

### 2. 訓練

```python
clf.fit(x_train, y_train)
```

### 3. 預測

```python
y_predict = clf.predict(x_test)
y_predict
```

```python
x0 = np.arange(-4, 4.2, 0.02)
y0 = np.arange(-1.5, 1.7, 0.02)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)

plt.contourf(xm, ym, Z, alpha=0.1)
plt.scatter(x_test[:,0], x_test[:,1], c=y_test)
```

# KMeans 基礎篇 (非監督式學習)

如果我們不知道答案，讓電腦自動分析吧！ (例如：客戶資料...)
原理類似於不停假設分類的中心，直到新分類的中心不變為止

```python
x = np.random.rand(100, 2)
plt.scatter(x[:,0], x[:,1], s=50)
```

### step 1. 打開 K-means() 學習機

```python
from sklearn.cluster import KMeans
```

```python
# n_cluster = 要分幾類
clf = KMeans(n_clusters=3)
```

### step 2. 訓練

```python
clf.fit(x)
```

### step 3. 預測

```python
clf.predict(x)
```

```python
# labels_ 是訓練完的答案
clf.labels_
```

```python
plt.scatter(x[:,0], x[:,1], c=clf.labels_)
```

```python
x0 = y0 = np.arange(-0.2, 1.2, 0.02)
xm, ym = np.meshgrid(x0, y0)

# c_ 是 collection 是一個 CClass ，後面要用中括號
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)

plt.contourf(xm, ym, Z, alpha=0.2)
plt.scatter(x[:,0], x[:,1], c=clf.labels_)
```

## make_blobs 生更像真實世界的數據

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
# n_samples 資料筆數
x, y = make_blobs(n_samples=500, centers=3,
                 n_features=2, random_state=0)
```

```python
# 產生了一些難判別的資料，夠真實了吧
plt.scatter(x[:,0], x[:,1], c=y)
```

### 來用 KMeans 玩玩

```python
plt.scatter(x[:,0], x[:,1])
```

### 1. 學習機

```python
clf = KMeans(n_clusters=3)
```

### 2. 訓練

```python
clf.fit(x)
```

### 3. 預測

```python
# KMeans 分類會有明顯邊界
plt.scatter(x[:,0], x[:,1], c=clf.labels_)

# 有時候加入 KMeans 會讓其他函數學習機預測能力變強!
```

# Mean-Shifts 連分幾類都不用說

### 1. 開學習機 

```python
from sklearn.cluster import MeanShift
```

```python
# 參數 bandwidth 可以決定群的大小
clf = MeanShift(bandwidth=1)
```

### 2. 訓練

```python
# MeanShift 會比較久，因為要判斷分幾類比較好
clf.fit(x)
```

### 3. 預測

```python
plt.scatter(x[:,0], x[:,1], c=clf.labels_)
```

```python
x0 = np.arange(-4.8, 5.2, 0.02)
y0 = np.arange(-2.5, 6.8, 0.02)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]

z = clf.predict(P)
Z = z.reshape(xm.shape)

plt.contourf(xm, ym, Z, alpha=0.3)
plt.scatter(x[:,0], x[:,1], c=clf.labels_)
```

# cross_val_score 為模型評分

原理：random_state 改變 cv 次，每次分數觀察

```python
plt.scatter(x[:,0], x[:,1], c=y)
```

```python
from sklearn.model_selection import cross_val_score
```

### SVC

```python
clf = SVC(gamma='auto')
```

```python
# cv= 切幾次
scores = cross_val_score(clf, x, y, cv=5)
scores
```

```python
# SVC 的平均分數
scores.mean()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
scores = cross_val_score(clf, x, y, cv=5)
scores
```

```python
scores.mean()
```

# 隨機森林 random forest - 總體學習(ensemble)的一種方法

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# n_estimators= 決策樹的樹木
clf = RandomForestClassifier(n_estimators=100)
```

```python
scores = cross_val_score(clf, x, y, cv=5)
```

```python
scores.mean()
```
