```python
#新增一個總和的欄位 df['欄位名稱'] = df[要總和的欄位的list].sum(1) ，參數設為 1 才會橫著加
df['總級分'] = df[['國文','英文','數學','自然','社會']].sum(1)
df.head()
```

```python
df['主科'] = df.數學*1.5 + df.英文
df.head()
```

```python
#照總級分由大到小排列
df.sort_values(by='總級分', ascending=False).head(20)
```

```python
#多項目進行排序，by的關鍵字參數指定為一個照重要程度排列的list
df.sort_values(by=['主科', '總級分'], ascending=False).head(20)
```

# 超棒的學習資源
[點我](https://github.com/justmarkham/pandas-videos)

```python
# UFO報告資料
dfu = pd.read_csv('http://bit.ly/uforeports')
dfu.head()
```

```python
#依項目集合值 df.groupby(項目)，計算次數 df.count()
dfu_state = dfu.groupby('State').count()
dfu_state
```

```python
#做排序，直接return值並不會改變本身
dfu_state.sort_values(by='Time', ascending=False)
```

```python
#如果要更動原列表，除了用原本變數去接，也能用參數 inplae
dfu_state.sort_values(by='Time', ascending=False, inplace=True)
dfu_state.head(10)
```

```python
#畫個圖，除了用 head()，更能用切片
#長條圖畫法 plot(kind='bar')
dfu_state[:10].Time.plot(kind='bar')
```

```python
#hist到底是啥小?
dfu_state[:10].Time.hist()
```

# pandas 初學困難點

```python
df = pd.DataFrame(np.random.randn(5,3),
                 index=range(1,6),
                 columns=list('ABC'))

df
```

### 尋找B欄大於0

```python
df[df.B>0]
```

### 如果B欄大於0，那C欄變成0

```python
df[df.B>0]['C'] = 0
```

```python
df
#上一個cell的動作創造了新的 dataframe，原本的沒動到
```

```python
#df.loc定位特定 index，要修改東西、找東西，用loc就對了
df.loc[2:3, 'B':'C']
```

```python
df.loc[2,'B']
```

```python
df.loc[2,'B'] = 1
```

```python
df
```

```python
df.loc[df.B>0, 'C'] = 0
```

```python
df
```
