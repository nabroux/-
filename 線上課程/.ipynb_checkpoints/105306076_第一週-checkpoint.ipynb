## 第二題

請以plot函式畫出臺北市近年來的高低溫分佈圖。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
highest = [19.1, 19.6, 22.1, 25.7, 29.2, 32.0, 34.3, 33.8, 31.1, 27.5, 24.2, 20.7]
lowest = [13.9, 14.2, 15.8, 19.0, 22.3, 24.6, 26.3, 26.1, 24.8, 22.3, 19.3, 15.6]
highest = np.array(highest)
lowest = np.array(lowest)
x = np.linspace(1,12,12)
plt.plot(x, highest)
plt.plot(x, lowest)
```
