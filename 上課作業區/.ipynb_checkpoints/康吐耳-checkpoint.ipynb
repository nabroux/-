```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.linspace(0, 10, 30)
y = np.linspace(0, 10, 30)
X, Y = np.meshgrid(x, y)
Z = np.random.randint(0, 5, 900)
Z = Z.reshape(X.shape)
plt.contourf(X, Y, Z, cmap='magma')
```
