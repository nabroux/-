```python
import random

gameOver = False
gesture = ['剪刀','石頭','布']
direction = ['上','下','左','右']

def chooseDirection(playerWin):
    print('選擇一個方向吧')
    playerDirection = direction[int(input('1:上 2:下 3:左 4:右  '))-1]
    comDirection = random.choice(direction)
    if playerWin:
        print('你手勢往', playerDirection, '，電腦的頭往', comDirection)
        if playerDirection == comDirection:
            print('你贏了！')
            return True
        else:
            print('可惜，再接再厲！')
    else:
        print('你頭往', playerDirection, '，電腦的手勢往', comDirection)
        if playerDirection == comDirection:
            print('你輸了！')
            return True
        else:
            print('恭喜，逃過一劫！')

print('來玩黑白猜吧！')
while True:
    player = int(input('\n1:剪刀 2:石頭 3:布  '))
    com = random.randint(1,3)
    print('你出了', gesture[player-1],'電腦出了', gesture[com-1])
    if player == com:
        print('平手，再一次')
    elif player == com-1:
        print('猜拳輸了')
        gameOver = chooseDirection(playerWin = False)
    elif player == 3 and com == 1:
        print('猜拳輸了')
        gameOver = chooseDirection(playerWin = False)
    else:
        print('猜拳贏了')
        gameOver = chooseDirection(playerWin = True)
        
    if gameOver == True:
        break
```
