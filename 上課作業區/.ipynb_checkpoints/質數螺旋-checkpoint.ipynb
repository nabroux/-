# 105306076 資管四 許雲輔 彩虹質數螺旋丸

```python
%matplotlib inline
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

```python
n = 1000
p = np.array([sp.prime(i) for i in range(1,n+1)])
```

```python
ax = plt.subplot(111, projection='polar')
ax.scatter(p, p, c=p, s=5, cmap='hsv', alpha=0.75)
```
